```python
import pandas as pd


new_column_names = [
    'customer',
    'state',
    'customer_lifetime_value',
    'response',
    'coverage',
    'education',
    'effective_to_date',
    'employment_status',
    'gender',
    'income',
    'location_code',
    'marital_status',
    'monthly_premium_auto',
    'months_since_last_claim',
    'months_since_policy_inception',
    'number_of_open_complaints',
    'number_of_policies',
    'policy_type',
    'policy',
    'renew_offer_type',
    'sales_channel',
    'total_claim_amount',
    'vehicle_class',
    'vehicle_size'
]


customer_df.columns = new_column_names
```

```python
customer_df.head()
```

```python
customer_df=customer_df.drop(columns=['customer'])
```

```python
customer_df.head()
```

```python
categorical_columns = customer_df.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_columns = customer_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
categorical_df = customer_df[categorical_columns]
numerical_df = customer_df[numerical_columns]
```

```python
categorical_df.head()
```

```python
numerical_df.head()
```

```python
for column in categorical_df.columns:
    plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
    sns.countplot(data=categorical_df, x=column, order=categorical_df[column].value_counts().index)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming categorical_df is your DataFrame containing categorical columns
# and numerical_df is your DataFrame containing numerical columns

# Combine categorical_df and numerical_df based on a shared index
combined_df = pd.concat([categorical_df, numerical_df], axis=1)

# Loop through each categorical column and create plots
for cat_column in categorical_df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=combined_df, x=cat_column, y='total_claim_amount')  # Change 'Total Claim Amount' to your target column
    plt.xticks(rotation=45)
    plt.title(f'{cat_column} vs Total Claim Amount')
    plt.show()
```

```python
# Loop through each categorical column and create plots
for cat_column in categorical_df.columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=combined_df, x=cat_column, y='total_claim_amount', estimator=sum)  # Change 'Total Claim Amount' to your target column
    plt.xticks(rotation=45)
    plt.title(f'{cat_column} vs Total Claim Amount')
    plt.show()
```

# Lab | Feature extraction

```python
categoricals = customer_df.select_dtypes(object)
categoricals.head()
```

```python
num_rows = len(categoricals.columns)
num_cols = 1 
```

```python
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 5 * num_rows), gridspec_kw={'hspace': 1})
fig.tight_layout()


for i, column in enumerate(categorical_df.columns):
    ax = axes[i] if num_rows > 1 else axes  
    sns.countplot(data=categorical_df, x=column, order=categorical_df[column].value_counts().index, ax=ax)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
    ax.set_title(f'Distribution of {column}')
    ax.set_xlabel('')
    ax.set_ylabel('Count')

plt.show()
```

```python
sns.set(style="whitegrid")
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
plt.figure(figsize=(12, 6))
sns.lineplot(data=customer_df, x='effective_to_date', y='total_claim_amount')
plt.title('Total Claim Amount Trend Over Time')
plt.xlabel('Effective To Date')
plt.ylabel('Total Claim Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
